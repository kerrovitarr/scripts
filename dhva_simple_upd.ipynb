```python
import numpy as np
from bokeh.io import output_notebook, push_notebook, show, curdoc
from bokeh.plotting import figure
from bokeh.layouts import column


curdoc().theme = 'dark_minimal'
```

```python
# Initialize Bokeh for notebook output
output_notebook()

# First Plot
# Generate data points
x = np.linspace(-3.14, 3.14, 100)
y = np.cos(x) + 1.30

# Create the first plot
p1 = figure(width=500, height=250, title="First Plot", x_axis_label='y', y_axis_label='x')
p1.line(y, x, color="blue", legend_label="y = cos(x) + 1.30")
p1.line(-y, x, color="red", legend_label="-y = -cos(x) + 1.30")

# Second Plot
# Generate data points
xx = np.linspace(0.3, 2.3, 100)
yy = 1.0 / (np.sin(np.arccos(xx - 1.3)) + 0.2)
yy[0] = 0.0
yy[99] = 0.0

# Create the second plot
p2 = figure(width=500, height=250, title="Second Plot", x_range=(0, 3))
p2.line(xx, yy, color="green")

# Third Plot
# Generate data points
x2 = np.linspace(0.01, 20, 100)
y2 = 0.4 * np.cos(x2 / 0.3) + np.cos(x2 / 2.3) + x2 / 13.0


# Create the third plot
p3 = figure(width=500, height=250, title="Third Plot")
p3.line(x2, y2, color="purple")

# Combine the plots
layout = column(p1, p2, p3)

# Display the plots
show(layout, notebook_handle=True)
```
